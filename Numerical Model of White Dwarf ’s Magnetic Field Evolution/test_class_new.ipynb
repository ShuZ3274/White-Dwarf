```python
import numpy as np
import matplotlib.pyplot as plt
from class_WD import White_Dwarf
```

```python
# Class White Dwarf, 1 total argument need to be specified;
# mass, put in the associated mass of the object
m09 = White_Dwarf(mass= 0.9)


# funcrion load data, 2 total argument need to be specified;
# directory: position of file
# filetype: default "general"
# "general": information of convection zone size, core soze, age. e.g: m09.csv
# "diffusivity": sample diffusivity file, e.g:file1.dat
# note: both tyoe need to be loaded at least once
m09.load_data("data/m09.csv")
m09.load_data("data/sigma/file1.dat", file_type="diffusivity")
m09.read_files("data/sigma")


# initialize the grid of white dwarf:
m09.initiate_radial_grid(npt = 512)


# evolve the field:
# intersteps: number of steps between each file, say between file1.dat and file2.dat
# r_out_frac: R_OUT/R_WD, defualt 0.2
R09evo_03 = m09.evolve_field("data/sigma", 0.5, inter_steps= 50, convection_boost= True, boost_factor= 100)


# testing with graph:
plt.plot(m09.r_grid, m09.R_evo[0])
plt.plot(m09.r_grid, R09evo_03[-1])
plt.yscale("log")
plt.show()
```
